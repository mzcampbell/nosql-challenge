## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
import json
import requests 
import pandas as pd 
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishment = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.
* Display the first document in the results using `pprint`.
* Convert the result to a Pandas DataFrame, print the number of rows in the DataFrame, and display the first 10 rows.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
match_query = {'$match':{'scores.Hygiene': 20}}
pipeline=[match_query]
results = list(establishment.aggregate(pipeline))
# Use count_documents to display the number of documents in the result
hygiene=establishment.count_documents({'scores.Hygiene': 20})
hygiene
# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(results_df))
# Display the first 10 rows of the DataFrame
results_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'AddressLine3': {'$in':['London']},'RatingValue': {'$gte':'4'}}

results_london = list(establishment.find(query))

# Use count_documents to display the number of documents in the result
rating_value=establishment.count_documents({'RatingValue': '4'})
rating_value
# Display the first document in the results using pprint
pprint(results_london[0])
```

```python
# Convert the result to a Pandas DataFrame
results_london_df = pd.DataFrame(results_london)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(results_london_df))
# Display the first 10 rows of the DataFrame
results_london_df.head(10)

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query = {'geocode.latitude': {'$gte': latitude-degree_search,'$lte': latitude+degree_search},
         'geocode.longitude': {'$gte': longitude-degree_search,'$lte': longitude+degree_search},
            'RatingValue':'5'}
                        
              
sort =  {'$sort': { 'scores.Hygiene': 1}}
pipelines= [query, sort]
results_long_lat= list(establishment.aggregate(pipeline))

# Print the results

print(results_long_lat[0:5])
```

```python
# Convert result to Pandas DataFrame
results_top5_df=pd.DataFrame(results_long_lat)
results_top5_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest


match_query = {'$match': {'scores.Hygiene': 0}}
group_query={'$group':{'_id':"$LocalAuthorityName",
                        'count':{'$sum':1}}}             
sort_values = {'$sort': { 'count':-1, '_id':1}}

# Put the pipeline together
pipeline = [match_query,group_query,sort_values]
result_hy = list(establishment.aggregate(pipeline))
# Print the number of documents in the result
print("Rows in DataFrame: ", len(result_hy))
# Display the first document in the results using pprint
pprint(result_hy[0])
```

```python
# Convert the result to a Pandas DataFrame
results_hygiene_df=pd.DataFrame(result_hy)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(results_hygiene_df))
# Display the first 10 rows of the DataFrame
results_hygiene_df.head(10)
```
